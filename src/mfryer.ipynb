```python
import pandas as pd
from process_transcripts import process_transcripts
from transcript_stats import get_transcript_stats
```

# Process the Transcripts


```python
process_transcripts()
```

# Get the Statistics


```python
df = get_transcript_stats()
```

# Calc Summary Stats


## By Metric


```python
summary_by_metric = (
    df[
        [
            "BrexitRisk",
            "BrexitExposure",
            "BrexitSentiment",
            "NonBrexitRisk",
            "NonBrexitSentiment",
        ]
    ]
    .agg(["mean", "std", "median"])
    .round(4)
)
summary_by_metric.transpose()
```

## By Ticker


```python
summary_by_ticker = (
    df.groupby("ticker")[
        [
            "BrexitRisk",
            "BrexitExposure",
            "BrexitSentiment",
            "NonBrexitRisk",
            "NonBrexitSentiment",
        ]
    ]
    .agg(["mean", "std", "median"])
    .round(4)
)
summary_by_ticker
```

## By Year


```python
summary_by_year = (
    df.groupby("year")[
        [
            "BrexitRisk",
            "BrexitExposure",
            "BrexitSentiment",
            "NonBrexitRisk",
            "NonBrexitSentiment",
        ]
    ]
    .agg(["mean", "std", "median"])
    .round(4)
)
summary_by_year
```
